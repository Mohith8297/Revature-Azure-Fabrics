```python
''' 1 Write Products into CSV File'''

import csv

products = [
    [101, "Laptop", "Electronics", 75000, 10],
    [102, "Mouse", "Accessories", 800, 50],
    [103, "Keyboard", "Accessories", 1500, 30]
]

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])
    writer.writerows(products)
```

```python
'''2 Read Data from CSV File'''

with open("products.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
'''3 Delete Product (id == 103)'''

rows = []
with open("products.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        if row[0] != "103":  # skip product with id 103
            rows.append(row)

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)
```

```python
'''4 Display Products with Price > 5000'''
with open("products.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        if int(row["Price"]) > 5000:
            print(row)
```

```python
'''5 Write Products into JSON File'''

import json

products_json = [
    {"id": 101, "name": "Laptop", "category": "Electronics", "price": 75000, "quantity": 10},
    {"id": 102, "name": "Mouse", "category": "Accessories", "price": 800, "quantity": 50},
    {"id": 103, "name": "Keyboard", "category": "Accessories", "price": 1500, "quantity": 30},
    {"id": 104, "name": "Mobile", "category": "Electronics", "price": 25000, "quantity": 15},
    {"id": 105, "name": "Headphones", "category": "Accessories", "price": 2000, "quantity": 20}
]

with open("products.json", "w") as file:
    json.dump(products_json, file, indent=4)
```

```python
'''6 Read Data from JSON File'''
with open("products.json", "r") as file:
    data = json.load(file)
    for product in data:
        print(product)
```

```python
'''7 Delete Product (id == 102)'''

with open("products.json", "r") as file:
    data = json.load(file)

data = [p for p in data if p["id"] != 102]

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)
```

```python
'''8 Update Product Quantity'''

with open("products.json", "r") as file:
    data = json.load(file)

for p in data:
    if p["name"] == "Laptop":
        p["quantity"] = 25

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)
```

```python
''' 9 Display Electronics Products Only'''
with open("products.json", "r") as file:
    data = json.load(file)

for p in data:
    if p["category"] == "Electronics":
        print(p)
```

```python
'''10  Append New Product to CSV'''

new_product = [106, "Monitor", "Electronics", 12000, 12]

with open("products.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(new_product)
```

```python
'''11 Append New Product to JSON'''

with open("products.json", "r") as file:
    data = json.load(file)

data.append({"id": 106, "name": "Monitor", "category": "Electronics", "price": 12000, "quantity": 12})

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)
```
