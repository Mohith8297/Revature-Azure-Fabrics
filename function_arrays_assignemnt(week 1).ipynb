```python

'''Using the list numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45] provide python solutions for the below problem statements.
1. Array Sum
Write a function that accepts a list of numbers and returns the sum of all elements.'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

def array_sum(arr):
    return sum(arr)
print("Sum:", array_sum(numbers))



```

```python
'''2. Maximum Element
Write a function to find the largest element in a list without using max().'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

def max_element(arr):
    max_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
    return max_val

print("Max:", max_element(numbers))



```

```python
'''3 Even Numbers Counter
  Write a function that counts how many even numbers are present in a list.'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]

def count_even(arr):
    return sum(1 for num in arr if num % 2 == 0)

print("Even Count:", count_even(numbers))

```

```python
'''4 Reverse Array
Write a function to reverse the elements of a list'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def reverse_array(arr):
    return arr[::-1]

print("Reversed:", reverse_array(numbers))
```

```python
'''5 Search Element
Write a function to search for a given element in a list and return its index.'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def search_element(arr, target):
    for i, num in enumerate(arr):
        if num == target:
            return i
    return -1

print("Index of 78:", search_element(numbers, 78))

```

```python
'''6 Palindrome List
Write a function to check whether a list reads the same forward and backward'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def is_palindrome(arr):
    return arr == arr[::-1]

print("Palindrome:", is_palindrome(numbers))
```

```python
'''7. Second Largest Number
Write a function to find the second largest number in a list'''


numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def second_largest(arr):
    unique_nums = list(set(arr))
    unique_nums.sort()
    return unique_nums[-2]

print("Second Largest:", second_largest(numbers))
```

```python
'''8 Duplicate Elements
Write a function that identifies duplicate elements in a list.'''


numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def find_duplicates(arr):
    seen = set()
    duplicates = set()
    for num in arr:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)
    return list(duplicates)

print("Duplicates:", find_duplicates(numbers))

```

```python
'''9 Function for Student Marks
Create a function that accepts student marks in a list and returns average, highest, and lowest marks.'''

numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def student_marks_analysis(marks):
    avg = sum(marks) / len(marks)
    highest = max(marks)
    lowest = min(marks)
    return avg, highest, lowest

marks = [88, 76, 92, 85, 69]
avg, high, low = student_marks_analysis(marks)
print("Average:", avg, "Highest:", high, "Lowest:", low)

```
