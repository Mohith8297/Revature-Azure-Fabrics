```python
import csv

products = [
    [101, "Laptop", "Electronics", 75000, 10],
    [102, "Mouse", "Accessories", 800, 50],
    [103, "Keyboard", "Accessories", 1500, 30]
]

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])

    writer.writerows(products)

print("CSV File Created Successfully")
```

Problem 2 — Read Data from CSV File
Read all products from CSV file and display each row.


```python
import csv

updated_products = []

with open("products.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        updated_products.append(row)

header = updated_products[0]

filtered_products = [header]

for row in updated_products[1:]:
    if row[0] != "103":
        filtered_products.append(row)

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(filtered_products)

print("Product 103 Deleted Successfully")
```

Problem 3 — Delete Product from CSV File
Delete product where id == 103 and rewrite updated data into CSV file.


```python
import csv

with open("products.csv", "r") as file:
    reader = csv.reader(file)

    next(reader)

    for row in reader:
        if int(row[3]) > 5000:
            print(row)
```

1.Create a line chart showing medicine distribution over months.


```python
import pandas as pd
import matplotlib.pyplot as plt

medical_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Medicine_Distributed': [1200, 1500, 1800, 1700, 2100, 2500],
    'Hospital_Orders': [30, 45, 50, 48, 60, 75],
    'Delivery_Time': [5, 4, 6, 5, 3, 2],
    'Revenue': [50000, 65000, 72000, 70000, 85000, 95000]
}

df = pd.DataFrame(medical_data)

plt.plot(df['Month'],
         df['Revenue'],
         marker='o')

plt.title("Monthly Revenue Growth")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.savefig("revenue_growth.png")
plt.show()
```

```python
plt.plot(df['Month'],
         df['Medicine_Distributed'],
         marker='o',
         label='Medicine Distributed')

plt.plot(df['Month'],
         df['Hospital_Orders'],
         marker='s',
         label='Hospital Orders')

plt.title("Medicine Distributed vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.grid(True)

plt.savefig("multiple_line_chart.png")
plt.show()
```

14.Create a pie chart for revenue contribution by month.


```python

highest = df['Revenue'].idxmax()

explode = [0, 0, 0, 0, 0, 0]
explode[highest] = 0.1

plt.pie(df['Revenue'],
        labels=df['Month'],
        autopct='%1.1f%%',
        explode=explode)

plt.title("Revenue Contribution by Month")

plt.savefig("revenue_explode_pie_chart.png")
plt.show()
```
