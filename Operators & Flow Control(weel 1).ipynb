```python
"""Q12. Simulate payment retry logic where the system retries payment a maximum of 3 times if payment fails."""

max_retry = 3
attempts = 0
while attempts<max_retry:
  attempts += 1
  print("Payment attempt:",attempts)
  print("Payment Failed Retry")
print("Payment failed after 3 attempts.")
```

```python
"""Q13. Write a program that keeps checking order status until status becomes 'Delivered' or maximum 5 attempts are reached."""

attempt = 0
max_attempts = 5
order_status = ""
while attempt < max_attempts:
  order_status = input("Enter the Order Status:")
  attempt+=1
  if order_status == "Delivered":
    print("Order delivered successfully")
    break
  else:
    print("Checking again aattempt:",attempt)
if order_status != "Delivered":
  print("Maximum attempts reached and order not delivered yet")
```

```python
"""Q14. Inventory restock check: Keep checking stock level every iteration and stop when stock becomes greater than 100."""


stock = 50
while stock <= 100:
  print("Current Stock: ",stock)
  stock = int(input("Enter the new stock level:"))
print('Stock is sufficient now:',stock)

```

```python
"""Q15. You are given a list of orders with order_id and amount.
- Skip cancelled orders
- Classify remaining orders as Premium or Regular
- Count total premium orders"""

orders = [
    {"order_id": 101, "amount": 8000, "status": "Delivered"},
    {"order_id": 102, "amount": 15000, "status": "Delivered"},
    {"order_id": 103, "amount": 12000, "status": "Cancelled"},
    {"order_id": 104, "amount": 5000, "status": "Delivered"},
    {"order_id": 105, "amount": 20000, "status": "Delivered"}
]

premium_count = 0  # counter for premium orders

for order in orders:
    if order["status"] == "Cancelled":
        continue  # skip cancelled orders

    if order["amount"] > 10000:
        print("Order", order["order_id"], "is Premium")
        premium_count += 1
    else:
        print("Order", order["order_id"], "is Regular")

print("Total Premium Orders:", premium_count)
```
